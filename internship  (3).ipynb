# Preprocessing

```python
#checking the null values 
data.isna().sum()
```

* only Date of termination and MangerID hve the null values and we fill the null values

```python
data['DateofTermination'].fillna('Still empoloyd', inplace=True)
data['ManagerID'].fillna(0, inplace=True)
```

```python
data.isna().sum()
```

```python
data.nunique()
```

* We will drop Employee_Name  and EmpID  columns, because they have no effect on the data

```python
#checking the outliers
numcols=data.dtypes[(data.dtypes=='int64') | (data.dtypes=='float64')].index
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for x in numcols:
    sns.boxplot(data[x],orient='h')
    plt.title(x)
    plt.show()
```

* for some columns Outliers can be very informative about data. So we drop these columns from outlier fixing.

```python
numcols=numcols.drop(['PerfScoreID','FromDiversityJobFairID','SpecialProjectsCount', 'DaysLateLast30','DeptID'])
```

```python
for x in numcols:
    print(x)
    descr = data[x].describe()
    IQR = descr['75%'] - descr['25%']
    upper_limit =  descr['75%'] + 1.5 * IQR
    lower_limit =  descr['25%'] - 1.5 * IQR  
    #outlier replacement
    data[x][data[x]>upper_limit] = upper_limit
    data[x][data[x]<lower_limit] = lower_limit
```

```python
for x in numcols:
    sns.boxplot(data[x],orient='h')
    plt.title(x)
    plt.show()
```

```python
#label encoder 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['Employee_Name']  = le.fit_transform(data['Employee_Name'])
data['State']  = le.fit_transform(data['State'])
data['Position']  = le.fit_transform(data['Position'])
data['DOB']  = le.fit_transform(data['DOB'])
data['Sex']  = le.fit_transform(data['Sex'])
data['MaritalDesc']  = le.fit_transform(data['MaritalDesc'])
data['CitizenDesc']  = le.fit_transform(data['CitizenDesc'])
data['HispanicLatino']  = le.fit_transform(data['HispanicLatino'])
data['RaceDesc']  = le.fit_transform(data['RaceDesc'])
data['DateofHire']  = le.fit_transform(data['DateofHire'])
data['DateofTermination']  = le.fit_transform(data['DateofTermination'])
data['TermReason']  = le.fit_transform(data['TermReason'])
data['EmploymentStatus']  = le.fit_transform(data['EmploymentStatus'])
data['Department']  = le.fit_transform(data['Department'])
data['ManagerName']  = le.fit_transform(data['ManagerName'])
data['ManagerID']  = le.fit_transform(data['ManagerID'])
data['RecruitmentSource']  = le.fit_transform(data['RecruitmentSource'])
data['PerformanceScore']  = le.fit_transform(data['PerformanceScore'])
data['LastPerformanceReview_Date']  = le.fit_transform(data['LastPerformanceReview_Date'])
data.head(30)
```

```python
data=data[[  'MarriedID', 'MaritalStatusID', 'GenderID','EmpStatusID', 'DeptID', 'PerfScoreID', 'FromDiversityJobFairID', 'Salary', 'Termd', 'PositionID', 'Position', 'State', 'Zip', 'DOB','MaritalDesc', 'CitizenDesc', 'HispanicLatino', 'RaceDesc', 'DateofHire', 'DateofTermination', 'TermReason', 'Department', 'ManagerName', 'ManagerID', 'RecruitmentSource','EngagementSurvey', 'EmpSatisfaction','SpecialProjectsCount', 'LastPerformanceReview_Date', 'DaysLateLast30','Absences' ]]
data
```

```python
from sklearn import preprocessing
min_max=preprocessing.MinMaxScaler(feature_range=(0,1))
x=min_max.fit_transform(data)
x=pd.DataFrame(x)
x.describe()
```

```python
#Splitting the dataset for vehicle_choice  prediction
x=data.loc[:, data.columns != 'Salary']
y=data['Salary']
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 42)
```

* Here the variable is continuous so we use regressor model

```python
# linear regression
from sklearn import linear_model
lr = linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
from sklearn.metrics import mean_squared_error,r2_score
print('MSE is:',mean_squared_error(y_test,y_pred))
print('r2 score is:',r2_score(y_test,y_pred))
```

```python
# lasso regressor
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(x_train,y_train)
y_pred=ls.predict(x_test)
print('MSE is:',mean_squared_error(y_test,y_pred))
print('r2 score is:',r2_score(y_test,y_pred))
```

```python
# decisiontree regressor
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred=dtr.predict(x_test)
print('MSE is:',mean_squared_error(y_test,y_pred))
print('r2 score is:',r2_score(y_test,y_pred))
```

```python
# randomforest regressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
print('MSE is:',mean_squared_error(y_test,y_pred))
print('r2 score is:',r2_score(y_test,y_pred))
```

```python
# fine tuning
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [2,4]
min_samples_split = [2, 5]
min_samples_leaf = [1, 2]
bootstrap = [True, False]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
rf_model = RandomForestRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
rf_grid = GridSearchCV(estimator = rf_model, param_grid = param_grid, cv = 10, verbose=2, n_jobs = 4)
```

```python
rf_grid.fit(x_train,y_train)
```
